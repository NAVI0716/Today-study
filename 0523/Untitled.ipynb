```python
data=pd.read_csv('경상북도_지역별_전통시장_현황.csv',encoding='EUC-KR')
data
```

```python
s_w=['시장','상가']
```

```python
from konlpy.tag import Okt
k_t=Okt()
from tqdm import tqdm 
X_data=[]
for i in tqdm(data['시장명']):
    k_t_m=k_t.morphs(i)
    end_d=[w for w in k_t_m if not w in s_w]
    X_data.append(''.join(end_d))
X_data
```

```python
y_data=data['시장 유형'].replace(['정기시장','상설시장'],[0,1])
```

```python
y_data
```

```python
import numpy as np
y_data=np.array(y_data)
```

```python
X_data=np.array(X_data)
```

```python
from sklearn.model_selection import train_test_split
x_data,test_x,y_data,test_y = train_test_split(X_data,y_data,test_size=0.3,random_state=0)
```
