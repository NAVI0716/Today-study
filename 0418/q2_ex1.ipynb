```python
from wordcloud import WordCloud#워드클라우드모듈
import matplotlib.pyplot as plt#그래프모듈
from collections import Counter#횟수적용모듈
from konlpy.tag import Okt
```

파일열기

```python
with open('ex1_data.txt','r',encoding='utf-8') as f:
    text=f.read()
text
```

```python
ok_t=Okt()
data1=ok_t.nouns(text)#명사추출
data1
```

```python
data2=Counter(i for i in data1 if len(i)>1)
data2
```

```python
data3=Counter(data2)
data3
```

```python
wc=WordCloud(font_path='C:\Windows\Fonts\malgun.ttf',max_words=100)
g_data=wc.generate_from_frequencies(data3)
plt.figure()
plt.imshow(g_data)
plt.axis('off')
plt.show
```
