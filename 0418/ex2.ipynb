```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#matplotlib.rcParams['font.family']='Malgun Gothic'
matplotlib.rc('font',family='Malgun Gothic')#동일한 의미
#matplotlib.rcParams['axes.unicode_minus']=False
matplotlib.rc('axes',unicode_minus=False)
```

```python
붓꽃data=sns.load_dataset('iris')
타이타닉data=sns.load_dataset('titanic')
팁data=sns.load_dataset('tips')
운송data=sns.load_dataset('flights')
```

```python
붓꽃data
# 타이타닉data
# 팁data
# 운송data
```

```python
x=붓꽃data.petal_length.values
sns.rugplot(x)#범위를 표현해줄수있는 그래프/분포도에 대한것을 선으로 표현
plt.title("붓꽃data중 꽃잎 길이 러그 플롯")
plt.show()
```

```python
sns.kdeplot(x)#부드럽게 얼마나 차지하는지 표현/ 밀도
plt.title("붓꽃data중 꽃잎 길이 커널밀도 플롯")
plt.show()
```

```python
sns.displot(x,kde=True,rug=True)
```
