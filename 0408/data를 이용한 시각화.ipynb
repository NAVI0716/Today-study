```python
x=['사망인원','부상인원','발생횟수']
b=[df['사망'],df['부상'],df['발생']]
사망=[]
부상=[]
발생=[]
for i in b[0]:
    사망.append(i)
for i in b[1]:
    부상.append(i)    
for i in b[1]:
    발생.append(i) 
사망=np.array(사망)
부상=np.array(부상)
발생=np.array(발생)
```

```python
x=['2016','2017','2018','2019','2020']
b=[df['사망'],df['부상'],df['발생']]
사망=[]
부상=[]
발생=[]
for i in b[0]:
    사망.append(i)
for i in b[1]:
    부상.append(i)    
for i in b[1]:
    발생.append(i) 
사망=np.array(사망)
부상=np.array(부상)
발생=np.array(발생)

plt.bar(x,사망)
plt.bar(x,부상,bottom=사망)
plt.bar(x,발생,bottom=부상+사망)

plt.show()
```
