```python
x=[1,2,3,4]
y=[4,5,6,7]
#plt.plot(x,y,'색상 마커 선종류')
plt.plot(x,y,'r*-')
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
#plt.plot(x,y,'색상 마커 선종류')
plt.plot(x,y,'r*-')
plt.xticks([1,2,3,4],labels=['data1','data2','data3','data4'])
plt.yticks([4,5,6,7],labels=['a','b','c','d'])
```

그래프영역표시

```python
x=[1,2,3,4]
y=[4,5,6,7]
#plt.fill_between(x의 좌표 범위, y의 좌표범위,명도)
#plt.fill#영역표시
#plt.fill_between#수평영역
#plt.fill_betweenx#수직영역
plt.plot(x,y)
plt.fill_between(x[1:3],y[1:3],alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]
plt.plot(x,y)
plt.plot(x,y1)
plt.fill_between(x[1:3],y[1:3],y1[1:3],alpha=0.5)
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]
plt.plot(x,y)
plt.plot(x,y1)
plt.fill_between([2.0,2.0,3.5,3.5],[2.0,6.0,6.0,3.0],alpha=0.5)#alpha투명도
```

축크기

```python
import numpy as np
x=np.linspace(-10,10,100)
y=x**3
plt.plot(x,y)
plt.xscale('symlog')
#plt.yscale('log')
```

```python
import numpy as np
x=np.linspace(0,10,100)
y=x**3
plt.plot(x,y)
```

```python
x=np.linspace(0,10,100)
y=np.exp(x)
y=x**3
plt.plot(x,y)
plt.yscale('log')
```

```python
x=[1,2,3,4]
y=[4,5,6,7]
y1=[1,3,4,5]
plt.plot(x,y)
plt.plot(x,y1)
#plt.grid(True,axis='x')
plt.grid(True,c='b',ls='--',axis='y',alpha=0.5)
```
