```python
from konlpy.corpus import kolaw
t_data=kolaw.open('constitution.txt')
```

```python
from soynlp import DoublespaceLineCorpus
t_data=DoublespaceLineCorpus(t_data.name,iter_sent=True)
from soynlp.word import WordExtractor
```

```python
for i in t_data:
    print(i)
    print()
```

```python
m=WordExtractor()
m.train(t_data)
```

```python
m_t=m.extract()
```

```python
단어_l=[]
ll=[]
tw2=Okt()
for i in t_data:
    sw_data=tw2.nouns(i)
    sw_data=[i for i in sw_data if len(i)>1]
    ll.append(sw_data)
ll
```

```python
from collections import Counter
단어_모음집=sum(ll,[])
len(단어_모음집)
결과_단어_모음집=Counter(단어_모음집)
결과_단어_모음집
정렬된_단어_모음=sorted(결과_단어_모음집.items(),key=lambda x:x[1],reverse=True)
len(정렬된_단어_모음)
```

```python
ck_t1=Tokenizer()
ck_t1.fit_on_texts(ll)
ck_t1.word_index
```

```python
encoded=ck_t1.texts_to_sequences(ll)
encoded
```

```python
ck_t1.word_counts
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences
end_data=pad_sequences(encoded,padding='post',truncating='post')
end_data
```
