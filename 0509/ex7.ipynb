data수집

```python
from konlpy.corpus import kolaw
t_data=kolaw.open('constitution.txt').read()
type(t_data)
```

전처리-(형태소 분석기 이용)

```python
from konlpy.tag import Okt
tw=Okt()#위의 객체가 태킹의 정보를 가지고있음
n_d=tw.nouns(t_data)#명사
type(n_d)
```

시각화 작업을 위한 LNTK Text 클래스 이용정리

```python
from  nltk import Text#단어를 취급해주는 클래스 #단어구성요소 패키지
ck_data=Text(n_d,name='kolaw')#딕셔너리파일 대체로 사용 곧바로 사용가능
type(ck_data)
```

정리된 내용을 이용하여 data분석

```python
from wordcloud import WordCloud
f =r'C:\Users\류성빈\AppData\Local\Microsoft\Windows\Fonts\Maplestory Bold.ttf'
wc=WordCloud(font_path=f,width=1000,height=600,background_color='white')
plt.imshow(wc.generate_from_frequencies(ck_data.vocab()))
plt.axis('off')
```
