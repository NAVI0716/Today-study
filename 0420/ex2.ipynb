```python
import numpy as np
import pandas as pd
data = {
    '이름':['길동','둘리',np.nan,'또치','희동',np.nan],
    '나이':[40,np.nan,15,np.nan,5,np.nan],
    '성별':['남자',np.nan,'여자','여자','남자',np.nan],
    '시험점수':[np.nan,20,80,10,2,np.nan]
}
df=pd.DataFrame(data)
df
```
