```python
from sklearn.datasets import *
import pandas as pd
l_data = load_boston()
data = pd.DataFrame(l_data.data,columns=l_data.feature_names)
pd.DataFrame(l_data.target,columns=['Y'])
data['Y']=l_data.target
data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(data,hue='Y')
```

```python
for i in range(3):
    k=list(l_data.feature_names[4*i:4+(4*i)])+['Y']
    sns.pairplot(vars=k,data=data)
    plt.show()
```

```python
for i in l_data.feature_names:
    plt.scatter(data['Y'],data[i])
    plt.xlabel('Y')
    plt.ylabel(i)
    plt.show()
```
