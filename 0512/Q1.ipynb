```python
from sklearn.datasets import *
import pandas as pd
data=load_breast_cancer()
c_data=pd.DataFrame(data.data,columns=data.feature_names)
data.target_names
sy = pd.Series(data.target,dtype='category')
sy = sy.cat.rename_categories(data.target_names)
c_data['Y']=sy
c_data
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(vars=['mean area','worst perimeter','worst area','worst symmetry'],data=c_data,hue='Y')
```

```python
X_data=c_data[['mean area','worst perimeter','worst area','worst symmetry']].values
Y_data=c_data['Y']
y_data=Y_data.replace(['malignant','benign'],[0,1])
y_data=y_data.values
y_data
```

```python
from tensorflow.keras.utils import to_categorical
y_data=to_categorical(y_data)
y_data
```

```python
from sklearn.model_selection import train_test_split
t_x,tt_x,t_y,tt_y=train_test_split(X_data,y_data,random_state=1)
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras import optimizers 
m = Sequential()
m.add(Dense(2, input_dim = 4,activation = 'sigmoid'))
m.compile(optimizer='adam',loss = 'binary_crossentropy'
          ,metrics=['binary_accuracy'])
hy = m.fit(t_x,t_y,epochs=200,validation_data=(tt_x,tt_y))
```

```python
x=range(1,len(hy.history['loss'])+1)
plt.plot(x,hy.history['loss'])
plt.plot(x,hy.history['val_loss'])
```
