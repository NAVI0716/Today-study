불용어 처리

```python
from nltk.corpus import stopwords
from string import punctuation
st=set(stopwords.words('english'))
p_ck=list(punctuation)
st.update(p_ck)
```

```python
f_l=[]
for i in X_data:
    l=[]
    
    for j in i.split():
        if j.strip().lower() not in st and j.strip().lower().isalpha():
            l.append(j.strip().lower())
    l_1=' '.join(l)

    f_l.append(l_1)
f_l
```

```python
df['v2']=f_l
df['v1']=data['v1'].values
df.v2.replace('',np.nan,inplace=True)
df.isna().sum()
```
