```python
from nltk.tokenize import sent_tokenize, word_tokenize
import gensim
from gensim.models import Word2Vec, FastText
s_data=open('data1.txt','r',encoding='utf-8').read()
s_data=s_data.replace('\n',' ')
s_data
d_l=[]
for 문장 in sent_tokenize(s_data):
    s_l=[]
    for 단어 in word_tokenize(문장):
        s_l.append(단어.lower())
    d_l.append(s_l)
d_l
```

```python
w_m1=Word2Vec(d_l,min_count=1,vector_size=100,window=5,sg=0)#Word2Vec(입력data,단어빈도수,임배딩된차원수,한묶음 단어개수,sg(0=CBOW,1=skip_gram))
```

```python
w_m1.wv.similarity('peter','wendy')
```
