## 평균제곱오차(Mean Square Error) 구하기
#### mse를 최소화하는 m, b를 구해 보기 바랍니다.

```python
data = [(1, 3), (2, 5)]
m = -1
b = 5
```

```python
m = 5
b = 4
```

```python
def get_mse(data, m, b): # y = wx + b
  n = len(data)
  squared_error = 0
  for x, y in data:
    # find predicted y
    y_hat = m*x+b
    squared_error += (y - y_hat)**2  # sigma

  # get average squared difference
  mse = squared_error / n
  return mse 
```

```python
print(get_mse(data, m, b))
```
