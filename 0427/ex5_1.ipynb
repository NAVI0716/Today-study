```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import load_iris
data = load_iris()
X =  data.data
#y =  data.target
n=range(1,10)
km_l=[KMeans(n_clusters=i) for i in n]
sc=[km_l[i].fit(X).inertia_ for i in range(len(km_l))]
plt.plot(n,sc)
plt.show()
data.target_names
```
