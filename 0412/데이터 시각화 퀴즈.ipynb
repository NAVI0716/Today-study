3. 스크린수 대비 매출액을 산출하여 1~10위까지를 시각화 하시오

```python
df3=df.copy()
df3['스크린수대비 매출액']=(df3['매출액']/df3['스크린수']).head(10)
df3=df3.sort_values(by='스크린수대비 매출액',ascending=False)
스크린매출액=df3['스크린수대비 매출액']
영화명=df3['영화명']
plt.title("스크린수대비 매출액")
plt.bar(영화명,스크린매출액)
plt.plot(영화명,스크린매출액,'r--')
plt.xticks(rotation=90)
plt.show()
```

4. 상영횟수 대비 관객수를 계산하여 1회 상영시 평균 관객수를 이용하여 1~10위 를 시각화 하시오

```python
df4=df.copy()
df4['상영횟수_대비_관객수']=(df4['관객수']//df4['상영횟수'])
df4=df4.sort_values(by='상영횟수_대비_관객수', ascending=False).head(10)
plt.bar(df4.영화명, df4.상영횟수_대비_관객수)
plt.plot(df4.영화명, df4.상영횟수_대비_관객수,'r--')
plt.xticks(rotation=90)
plt.show()
```

5. 관객수 대비 누적매출액을 계산하여 평균 1명당 영화 관람에 사용한 금액을 계산하여 1~10위까지의 순위를 시각화 하시오

```python
df5=df.copy()
df5['관객수_대비_누적매출액']=(df5['누적매출액']/df5['관객수'])
df5=df5.sort_values(by='관객수_대비_누적매출액', ascending=False).head(10)
plt.bar(df5.영화명, df5.관객수_대비_누적매출액)
plt.plot(df5.영화명, df5.관객수_대비_누적매출액,'r--')
plt.xticks(rotation=90)
plt.show()
```

6. 스크린수 대비 상영횟수를 계산하여 평균 1개의 스크린에서 상영한 횟수를 통한 1~10위 순위를 시각화 하시오

```python
df6=df.copy()
df6['스크린수_대비_상영횟수']=(df6['상영횟수']/df6['스크린수'])
df6=df6.sort_values(by='스크린수_대비_상영횟수', ascending=False).head(10)
plt.bar(df6.영화명, df6.스크린수_대비_상영횟수)
plt.plot(df6.영화명, df6.스크린수_대비_상영횟수,'r--')
plt.xticks(rotation=90)
plt.show()
```

7. 매출액 점유 기준 1~10위 를 선별하고 원그래프를 만드시오

```python
df7 = df.sort_values(by='매출액', ascending=False).head(10)
plt.pie(df7.매출액, labels=df7.영화명, autopct='%.2f%%')
plt.show()
```
