```python
import matplotlib.pyplot as plt
import numpy as np
import random

def data_gen(n,b,v):
    x = np.zeros(shape=(n,2))#비어있는 데이터 생성 zeros
    y = np.zeros(shape=n)
    
    for i in range(0,n):
        x[i][0] = 1
        x[i][1] = i
        y[i] = (i+b) + random.uniform(0,1)*v #x라고하는것들을 기준으로 학습을 진할때 대응하는 학습률의 역할을 한다. 
    return x,y
x,y=data_gen(100,25,10)#데이터 100개 ,
plt.plot(x[:,1],y,'o')#[:,1]1차원 값
```

```python
def g_descent(x,y,th,ap,m,n):
    xT=x.T#np.traspose(x), x.transpose(),x.T
    th_l=[]
    cost_l=[]
    for i in range(0,ns):#내적
        h = np.dot(x,th) #np.dot(x,th), x.dot(th)
        loss = h-y #오차 생성
        cost = np.sum(loss**2)/(2*m) #
        g = np.dot(xT,loss)/m #기울기요소 언급하기 위함
        th=th - ap*g #th세타 ap 알파 g 기울기 / 세타(w0 세타0, w1 세타1)는 x의 위치 /알파는 학습률 runnigrate/ m 얼마만큼의 데이터가 있냐 입력데이터 계수
        if i %250 == 0:#5000번인데 250번당 이니까 20개의 데이터 발생
            th_l.append(th)
        cost_l.append(cost)
    return th,np.array(th_l),cost_l
m,n = np.shape(x)
ns = 5000#반복수
ap = 0.0005
th = np.ones(n)
th,th_l,cost_l = g_descent(x,y,th,ap,m,ns)
y_p_step= np.dot(x,th_l.T)
plt.plot(x[:,1],y,'o')
for i in range(0,20,5):
    plt.plot(x[:,1],y_p_step[:,i],label=f"L_{i}")
plt.legend()
plt.show()   
 
```
