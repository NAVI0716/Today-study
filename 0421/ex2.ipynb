```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(-10,11,1)
fx=x**2
#y = x**2

plt.plot(x,fx)
```

```python
x = np.arange(-10,11,1)
fx=x**2

x_new = 10
d=[]#변화된 a기록
y=[]#y값기록
learng_rate = 0.1#하이퍼파라미터값
for i in range(100):
    x_old=x_new
    x_new=x_old-learng_rate*2*x_old
    d.append(x_new)
    y.append(x_new**2)
plt.plot(x,fx)#곡선에 대한내용
plt.scatter(d,y)
```

```python
x = np.arange(-10,11,1)
fx=x**2

x_new = 10
d=[]#변화된 a기록
y=[]#y값기록
learng_rate = 0.1#하이퍼파라미터값
for i in range(100):
    x_old=x_new
    x_new=x_old-learng_rate*2*x_old
    d.append(x_new)
    y.append(x_new**2)
plt.plot(x,fx)#곡선에 대한내용
plt.scatter(d,y)
plt.plot(d,y)
```

## mae
평균절대잔차<br><br>
예측값과 실제값 차이에 절대값하고 이를 더해 평균구함

```python
from sklearn.metrics import mean_absolute_error #절댓값오차 코드
y_true=[3,-0.5,2,7]#예측값
y_pred=[2.5,0.0,2,8]#실제값
#위 두가지로 오차계산
mean_absolute_error(y_true,y_pred)
```
